```python
import pandas as pd

# Assuming your DataFrame is named `df`

# Aggregate total runs scored and balls faced by each striker
player_stats = df.groupby('striker').agg(
    total_runs=pd.NamedAgg(column='runs_off_bat', aggfunc='sum'),
    total_balls=pd.NamedAgg(column='ball', aggfunc='count')
).reset_index()

# Calculate the strike rate
player_stats['strike_rate'] = (player_stats['total_runs'] / player_stats['total_balls']) * 100

# Display the player stats with strike rate
print(player_stats.head())
```

```python
import pandas as pd

# Assuming your DataFrame is named `df`

# Replace NaNs in extras columns with 0 and convert to integers
df['wides'] = df['wides'].fillna(0).astype(int)
df['noballs'] = df['noballs'].fillna(0).astype(int)
df['byes'] = df['byes'].fillna(0).astype(int)
df['legbyes'] = df['legbyes'].fillna(0).astype(int)

# Aggregate total runs conceded and total balls bowled by each bowler
runs_conceded = df.groupby('bowler').agg(
    total_runs_conceded=pd.NamedAgg(column='runs_off_bat', aggfunc='sum')
).reset_index()

# Add runs from extras to the runs conceded
runs_conceded['total_runs_conceded'] += df.groupby('bowler').agg(
    total_wides=pd.NamedAgg(column='wides', aggfunc='sum')
).reset_index()['total_wides']

runs_conceded['total_runs_conceded'] += df.groupby('bowler').agg(
    total_noballs=pd.NamedAgg(column='noballs', aggfunc='sum')
).reset_index()['total_noballs']

runs_conceded['total_runs_conceded'] += df.groupby('bowler').agg(
    total_byes=pd.NamedAgg(column='byes', aggfunc='sum')
).reset_index()['total_byes']

runs_conceded['total_runs_conceded'] += df.groupby('bowler').agg(
    total_legbyes=pd.NamedAgg(column='legbyes', aggfunc='sum')
).reset_index()['total_legbyes']

# Aggregate total balls bowled by each bowler
balls_bowled = df.groupby('bowler').agg(
    total_balls=pd.NamedAgg(column='ball', aggfunc='count')
).reset_index()

# Merge runs_conceded and balls_bowled DataFrames
bowler_stats = pd.merge(runs_conceded, balls_bowled, on='bowler')

# Convert total balls to overs
bowler_stats['total_overs'] = bowler_stats['total_balls'] / 6

# Calculate the economy rate
bowler_stats['economy_rate'] = bowler_stats['total_runs_conceded'] / bowler_stats['total_overs']

# Display the bowler stats with economy rate
print(bowler_stats.head())
```
