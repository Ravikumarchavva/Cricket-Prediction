```python
import pandas as pd
import requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
full_table=pd.DataFrame()
for i in range(1,4):
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;page={i};template=results;type=team"
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Raise an exception for HTTP errors
    # Read HTML with pandas
    tables = pd.read_html(response.text, flavor='bs4')  # or use flavor='html5lib'
    stats_table = tables[2]
    full_table=pd.concat([stats_table, full_table], axis=0, ignore_index=True)
```
